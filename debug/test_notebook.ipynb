```python
import matplotlib.pyplot as plt
import numpy as np 
from mpl_bsic import apply_bsic_style, export_figure, add_title_subtitle
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)

fig, ax = plt.subplots(2,2, figsize=(8,8))
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin(x)')

# fig.savefig('test.svg', dpi=1200, bbox_inches='tight')
apply_bsic_style(fig)
title = 'Sin and Cosin of X'
subtitle = 'Yu can see from the figure how the sin and cosin of x are just shifted versions of each other. We will now do some further analysis on this'
fig.subplots_adjust(top=0.9)
ann1, ann2 = add_title_subtitle(fig, title, subtitle)
# # txt = ax[0,0].text(0,ax[0,0].get_ylim()[1],subtitle, ha='left', va='center', wrap=True)
# width = ax[0,0].bbox.width * 2
# width = ax[0,0].get_tightbbox().width
# print(width)
# ann1._get_wrap_line_width = lambda : width
# 
# bbox = fig.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
# print(bbox.width * fig.dpi)
# fig.tight_layout()
# export_figure(fig, 'testfinal')
# fig.subplots_adjust(top=0.8)
# fig.savefig('test.png', bbox_inches='tight')    

fig.savefig('testone.png')
```
